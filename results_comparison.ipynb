```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/TA - Helmet Violation/Result/Training-Result/Model/GA-COCO/results.csv', sep=';')
df2 = pd.read_csv('/content/drive/MyDrive/TA - Helmet Violation/Result/Training-Result/Model/GA-OpenImages/results.csv', sep=';')
df3 = pd.read_csv('/content/drive/MyDrive/TA - Helmet Violation/Result/Training-Result/Model/Standar-COCO/results.csv', sep=';')
df4 = pd.read_csv('/content/drive/MyDrive/TA - Helmet Violation/Result/Training-Result/Model/Standar-OpenImages/results (1).csv', sep=';')

df1.head()
```

```python
fig, axs = plt.subplots(2, 5, figsize=(20, 10))

# Plot train losses
axs[0, 0].plot(df1['epoch'], df1['train/box_loss'], 'o-', label='GA-COCO', linewidth=2, markersize=5)
axs[0, 0].plot(df2['epoch'], df2['train/box_loss'], 'o-', label='GA-OpenImages', linewidth=2, markersize=5)
axs[0, 0].plot(df3['epoch'], df3['train/box_loss'], 'o-', label='Standar-COCO', linewidth=2, markersize=5)
axs[0, 0].plot(df4['epoch'], df4['train/box_loss'], 'o-', label='Standar-OpenImages', linewidth=2, markersize=5)
axs[0, 0].set_title('train/box_loss')

axs[0, 1].plot(df1['epoch'], df1['train/cls_loss'], 'o-', label='GA-COCO', linewidth=2, markersize=5)
axs[0, 1].plot(df2['epoch'], df2['train/cls_loss'], 'o-', label='GA-OpenImages', linewidth=2, markersize=5)
axs[0, 1].plot(df3['epoch'], df3['train/cls_loss'], 'o-', label='Standar-COCO', linewidth=2, markersize=5)
axs[0, 1].plot(df4['epoch'], df4['train/cls_loss'], 'o-', label='Standar-OpenImages', linewidth=2, markersize=5)
axs[0, 1].set_title('train/cls_loss')

axs[0, 2].plot(df1['epoch'], df1['train/dfl_loss'], 'o-', label='GA-COCO', linewidth=2, markersize=5)
axs[0, 2].plot(df2['epoch'], df2['train/dfl_loss'], 'o-', label='GA-OpenImages', linewidth=2, markersize=5)
axs[0, 2].plot(df3['epoch'], df3['train/dfl_loss'], 'o-', label='Standar-COCO', linewidth=2, markersize=5)
axs[0, 2].plot(df4['epoch'], df4['train/dfl_loss'], 'o-', label='Standar-OpenImages', linewidth=2, markersize=5)
axs[0, 2].set_title('train/dfl_loss')

# Plot metrics
axs[0, 3].plot(df1['epoch'], df1['metrics/precision(B)'], 'o-', label='GA-COCO', linewidth=2, markersize=5)
axs[0, 3].plot(df2['epoch'], df2['metrics/precision(B)'], 'o-', label='GA-OpenImages', linewidth=2, markersize=5)
axs[0, 3].plot(df3['epoch'], df3['metrics/precision(B)'], 'o-', label='Standar-COCO', linewidth=2, markersize=5)
axs[0, 3].plot(df4['epoch'], df4['metrics/precision(B)'], 'o-', label='Standar-OpenImages', linewidth=2, markersize=5)
axs[0, 3].set_title('metrics/precision(B)')

axs[0, 4].plot(df1['epoch'], df1['metrics/recall(B)'], 'o-', label='GA-COCO', linewidth=2, markersize=5)
axs[0, 4].plot(df2['epoch'], df2['metrics/recall(B)'], 'o-', label='GA-OpenImages', linewidth=2, markersize=5)
axs[0, 4].plot(df3['epoch'], df3['metrics/recall(B)'], 'o-', label='Standar-COCO', linewidth=2, markersize=5)
axs[0, 4].plot(df4['epoch'], df4['metrics/recall(B)'], 'o-', label='Standar-OpenImages', linewidth=2, markersize=5)
axs[0, 4].set_title('metrics/recall(B)')

axs[1, 3].plot(df1['epoch'], df1['metrics/mAP50(B)'], 'o-', label='GA-COCO', linewidth=2, markersize=5)
axs[1, 3].plot(df2['epoch'], df2['metrics/mAP50(B)'], 'o-', label='GA-OpenImages', linewidth=2, markersize=5)
axs[1, 3].plot(df3['epoch'], df3['metrics/mAP50(B)'], 'o-', label='Standar-COCO', linewidth=2, markersize=5)
axs[1, 3].plot(df4['epoch'], df4['metrics/mAP50(B)'], 'o-', label='Standar-OpenImages', linewidth=2, markersize=5)
axs[1, 3].set_title('metrics/mAP50(B)')

axs[1, 4].plot(df1['epoch'], df1['metrics/mAP50-95(B)'], 'o-', label='GA-COCO', linewidth=2, markersize=5)
axs[1, 4].plot(df2['epoch'], df2['metrics/mAP50-95(B)'], 'o-', label='GA-OpenImages', linewidth=2, markersize=5)
axs[1, 4].plot(df3['epoch'], df3['metrics/mAP50-95(B)'], 'o-', label='Standar-COCO', linewidth=2, markersize=5)
axs[1, 4].plot(df4['epoch'], df4['metrics/mAP50-95(B)'], 'o-', label='Standar-OpenImages', linewidth=2, markersize=5)
axs[1, 4].set_title('metrics/mAP50-95(B)')

# Plot validation losses
axs[1, 0].plot(df1['epoch'], df1['val/box_loss'], 'o-', label='GA-COCO', linewidth=2, markersize=5)
axs[1, 0].plot(df2['epoch'], df2['val/box_loss'], 'o-', label='GA-OpenImages', linewidth=2, markersize=5)
axs[1, 0].plot(df3['epoch'], df3['val/box_loss'], 'o-', label='Standar-COCO', linewidth=2, markersize=5)
axs[1, 0].plot(df4['epoch'], df4['val/box_loss'], 'o-', label='Standar-OpenImages', linewidth=2, markersize=5)
axs[1, 0].set_title('val/box_loss')

axs[1, 1].plot(df1['epoch'], df1['val/cls_loss'], 'o-', label='GA-COCO', linewidth=2, markersize=5)
axs[1, 1].plot(df2['epoch'], df2['val/cls_loss'], 'o-', label='GA-OpenImages', linewidth=2, markersize=5)
axs[1, 1].plot(df3['epoch'], df3['val/cls_loss'], 'o-', label='Standar-COCO', linewidth=2, markersize=5)
axs[1, 1].plot(df4['epoch'], df4['val/cls_loss'], 'o-', label='Standar-OpenImages', linewidth=2, markersize=5)
axs[1, 1].set_title('val/cls_loss')

axs[1, 2].plot(df1['epoch'], df1['val/dfl_loss'], 'o-', label='GA-COCO', linewidth=2, markersize=5)
axs[1, 2].plot(df2['epoch'], df2['val/dfl_loss'], 'o-', label='GA-OpenImages', linewidth=2, markersize=5)
axs[1, 2].plot(df3['epoch'], df3['val/dfl_loss'], 'o-', label='Standar-COCO', linewidth=2, markersize=5)
axs[1, 2].plot(df4['epoch'], df4['val/dfl_loss'], 'o-', label='Standar-OpenImages', linewidth=2, markersize=5)
axs[1, 2].set_title('val/dfl_loss')

plt.tight_layout()
plt.legend()
plt.show()
```
