```python
import os
import subprocess
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
from ultralytics import YOLO
import pandas as pd

model_standar_COCO = YOLO('/kaggle/input/model-yolov8/Standar-COCO/best.pt')
model_standar_OI = YOLO('/kaggle/input/model-yolov8/Standar-OpenImages/best (2).pt')
model_GA_COCO = YOLO('/kaggle/input/model-yolov8/GA-COCO/weights/best.pt')
model_GA_OI = YOLO('/kaggle/input/model-yolov8/GA-OpenImages/weights/best.pt')

# Define the directory containing the videos
video_directory = '/kaggle/input/dataset1-renamed'
```

```python
results_dict = {
    'Video': [],
    'Total_Speed_Standar_COCO': [],
    'Total_Speed_Standar_OI': [],
    'Total_Speed_GA_COCO': [],
    'Total_Speed_GA_OI': []
}


# Iterate over the videos in the directory
for i in range(1, 7): 
    video_file = f'({i}).mp4'
    video_path = os.path.join(video_directory, video_file)

    # Perform inference on the video
    results = model_standar_COCO(video_path, stream=False, conf=0.45)

    # Calculate total speed for each video
    total_speed = sum(sum(results[j].speed.values()) for j in range(len(results))) / len(results)

    # Append results to the dictionary
    results_dict['Video'].append(video_file)
    results_dict['Total_Speed_Standar_COCO'].append(total_speed)
```

```python
# Convert the results dictionary to a DataFrame
results_df = pd.DataFrame(results_dict)

# Round the total speed columns to 3 decimal places
results_df['Total_Speed_Standar_COCO'] = results_df['Total_Speed_Standar_COCO'].round(3)
results_df['Total_Speed_Standar_OI'] = results_df['Total_Speed_Standar_OI'].round(3)
results_df['Total_Speed_GA_COCO'] = results_df['Total_Speed_GA_COCO'].round(3)
results_df['Total_Speed_GA_OI'] = results_df['Total_Speed_GA_OI'].round(3)

# Calculate the average for each column and create a new DataFrame for the average row
average_row = pd.DataFrame({
    'Video': ['Average'],
    'Total_Speed_Standar_COCO': [results_df['Total_Speed_Standar_COCO'].mean()],
    'Total_Speed_Standar_OI': [results_df['Total_Speed_Standar_OI'].mean()],
    'Total_Speed_GA_COCO': [results_df['Total_Speed_GA_COCO'].mean()],
    'Total_Speed_GA_OI': [results_df['Total_Speed_GA_OI'].mean()]
})

# Concatenate the original DataFrame with the average row DataFrame
results_df = pd.concat([results_df, average_row], ignore_index=True)

results_df
```

```python
def convert_and_compress(video_path, output_format="mp4"):
  """
  Converts a video to the specified output format and potentially compresses it.

  Args:
    video_path: Path to the video file.
    output_format: The desired output format (e.g., "mp4", "avi").
  """
  # Get filename and extension
  filename, extension = os.path.splitext(video_path)

  # Check if video is .avi
  if extension.lower() != ".avi":
    print(f"Skipping {video_path}, not an AVI file.")
    return

  # Create output filename with new extension
  output_path = f"{filename}.{output_format}"

  # Use ffmpeg to convert and compress (replace with your preferred encoding options)
  command = ["ffmpeg", "-i", video_path, "-c:v", "libx264", "-c:a", "aac", output_path]
  subprocess.run(command, check=True)
  print(f"Converted {video_path} to {output_path}")

# Get all files in the directory
files = os.listdir("/kaggle/working/runs/detect/predict")

# Loop through files and convert/compress .avi files
for file in files:
  video_path = os.path.join("/kaggle/working/runs/detect/predict", file)
  convert_and_compress(video_path)
```
