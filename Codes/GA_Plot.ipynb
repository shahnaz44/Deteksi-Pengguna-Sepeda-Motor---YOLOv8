```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

# COCO Weight

```python
base_dir = '/content/drive/MyDrive/TA - Helmet Violation/Result/Training-Result/Genetic Algorithm/Tunes-COCO'
combined_df = pd.DataFrame()
```

```python
for i in range(1, 13):
    folder_path = os.path.join(base_dir, str(i))
    file_path = os.path.join(folder_path, 'tune_results.csv')

    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Optionally add a column to identify the source folder
    df['folder'] = i

    # Append to the combined DataFrame
    combined_df = pd.concat([combined_df, df], ignore_index=True)
```

```python
combined_df
```

```python
combined_df['generation'] = range(1, len(combined_df) + 1)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(combined_df['generation'], combined_df['fitness'], alpha=0.5)
plt.title('Fitness vs Generasi')
plt.xlabel('Generasi')
plt.ylabel('Fitness')
plt.grid(True)
plt.show()
```

```python
combined_df.sort_values(by = 'fitness')
```
